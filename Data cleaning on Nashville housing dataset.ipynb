DATA CEANING

```python
# Finding missing values
data.isnull().any()
```

```python
data1=data.dropna()
data1.isnull().any()
```

```python
# Removing Duplicate
data.duplicated().any()
# no duplicates but we deal with duplicates like this
data2=data1.drop_duplicates()
data2.duplicated().any()
```

```python
data2['SoldAsVacant']
```

```python
data2['SoldAsVacant'].unique()
```

```python
# Replacing 'N' with 'No' and 'Y' with 'Yes'
data2['SoldAsVacant'] = data2['SoldAsVacant'].replace('N', 'No')
data2['SoldAsVacant'] = data2['SoldAsVacant'].replace('Y', 'Yes')
```
